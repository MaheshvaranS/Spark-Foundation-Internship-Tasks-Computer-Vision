```python
from imageai.Detection import ObjectDetection
import os
import time
```

```python
from tensorflow.python.framework.ops import disable_eager_execution
disable_eager_execution()
```

```python
yolo_obj = ObjectDetection()
yolo_obj.setModelTypeAsYOLOv3()
yolo_obj.setModelPath("/content/yolo.h5")
yolo_obj.loadModel()
from PIL import Image
Image.open("/content/sample_input.jpg")

start = time.time()
detections = yolo_obj.detectObjectsFromImage(input_image = ("/content/sample_input.jpg"), output_image_path = ("/content/out_img1.jpg"))
print('sec',time.time() - start)

for objects in detections:
    print(objects["name"] , " : " , objects["percentage_probability"] )

Image.open("/content/out_img1.jpg")
```
